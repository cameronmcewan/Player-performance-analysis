```python
import requests

standings_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
data = requests.get(standings_url)

# Check for a successful response
if data.status_code == 200:
    html = data.content
    print('Successfully retrieved page')
else:
    print(f'Failed to retrieve page with error code: {data.status_code}')
```

```python
from bs4 import BeautifulSoup

# soup = BeautifulSoup(html, 'html.parser')
soup = BeautifulSoup(data.text)
```

```python
# soup.select uses css selectors
standings_table = soup.select('table.stats_table')[0]
```

```python
# .find_all only searches for tags
links = standings_table.find_all('a')
links = [l.get("href") for l in links]
links = [l for l in links if '/squads/' in l]
links
# these are relative links
```

```python
# create full (absolute) links
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

```python
spurs_url = team_urls[0]
```

```python
data = requests.get(spurs_url)
data
```

```python
import pandas as pd

#pd.read_html scans all tables on the page, searching for a match with the string specified
matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
matches.head()
```

# Get Shooting Stats
- Initialise new BeautifulSoup instance
- Find all links on page
- Find "href"
- Find links containing Shooting Stats

```python
soup = BeautifulSoup(data.text)
links = soup.find_all('a')
links = [l.get("href") for l in links]
links = [l for l in links if l and 'all_comps/shooting/' in l]
links
```

```python
data = requests.get(f"https://fbref.com{links[0]}")
# data.text
```

```python
# Match on the string "Shooting"
shooting = pd.read_html(data.text, match="Shooting")[0]
```

```python
shooting.head()
```

```python
# The multi level index isn't helpful, so drop 1 index level
shooting.columns = shooting.columns.droplevel()
```
